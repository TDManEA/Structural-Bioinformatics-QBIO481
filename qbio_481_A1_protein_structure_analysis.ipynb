1A. Interactions between Amino Acid Side Chains
    a. Lysine and Glutamic Acid
    Interaction: Electrostatic (ionic) interaction.
    Explanation: Lysine has a positively charged amine group (-NH₃⁺) at physiological pH, whereas glutamic acid has a negatively charged carboxylate group (-COO⁻). Therefore, the two opposite charges attract each other, resulting in a strong ionic interaction(s).

    b. Asparagine and Serine
    Interaction: Hydrogen bonding.
    Explanation: Both asparagine and serine have residues with polar groups. Asparagine has an amide group (-CONH₂), and serine has a hydroxyl group (-OH). These groups can form hydrogen bonds with each other—due to their propensity to donate and accept hydrogen atoms—and are hydrophilic, suggesting that both amino acids can participate in bonding with water or other polar molecules. 

    c. Tryptophan and Tyrosine
    Interaction: Pi stacking and hydrogen bonding.
    Explanation: Both tryptophan and tyrosine have aromatic rings in their side chains, which can participate in π-π stacking interactions (non-covalent interactions between aromatic rings). Additionally, tyrosine has a hydroxyl group that can participate in hydrogen bonding, which is significantly more powerful an interaction: between the two. However, it is important to not that the stabilizing effect of π stacking yields a synergistic stability—where both mechanisms work together. 

1B. Hydrogen Bonding Configurations
    a. α-Helices:
    Hydrogen bonds form between the carbonyl oxygen (C=O) of residue i and the
        amide hydrogen (N-H) of residue i+4. This pattern gives rise to a stable helical structure.

    b. 3₁₀-Helices:
    Hydrogen bonds form between the carbonyl oxygen of residue i and the
        amide hydrogen of residue i+3. This creates a tighter, more compact helical turn compared to an α-helix.

    c. π-Helices:
    Hydrogen bonds form between the carbonyl oxygen of residue i and the 
        amide hydrogen of residue i+5. This structure is more extended compared to the α-helix and is less common.

    d. β-Sheets:
    Hydrogen bonds form between the carbonyl oxygen of one strand and 
        the amide hydrogen of an adjacent, parallel or antiparallel strand. These interactions stabilize the sheet-like arrangement.

1C. Antiparallel vs. Parallel β-Sheets
    Antiparallel β-sheets are more commonly found because their hydrogen bonds are more linear and stable. In antiparallel sheets, the strands run in opposite directions, allowing hydrogen bonds to form linear bond angles, minimizing the distance between the strands (2.8 - 3.2 A), which increases stability. In contrast, parallel β-sheets (3.2 - 3.5 A)have strands running in the same direction, causing the hydrogen bonds to be angled, resulting in weaker interactions—and less stability, overall.

```python
### Part 2A
print('\n ### Part 2A \n')
# Part A: Total Interaction Energy
# Step 1: Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Step 2: Define sample parameters
q1 = 2          # Charge of ion 1
q2 = -1         # Charge of ion 2
A = 0.2         # Parameter for Lennard-Jones potential
b = 1.5         # Parameter for Lennard-Jones potential

# Step 3: Define the range for distance r
r = np.linspace(1.2, 3, 100)  # Distance from 1.2 to 3

# Step 4: Calculate the Coulomb potential
def V_coulomb(r, q1, q2):
    return (q1 * q2) / r

# Step 5: Calculate the Lennard-Jones potential
def V_lennard_jones(r, A, b):
    return A * ((b / r) ** 12 - 2 * (b / r) ** 6)

# Step 6: Calculate the total potential
def V_total(r, q1, q2, A, b):
    return V_coulomb(r, q1, q2) + V_lennard_jones(r, A, b)

# Step 7: Compute potential values
V_coulomb_values = V_coulomb(r, q1, q2)
V_lennard_jones_values = V_lennard_jones(r, A, b)
V_total_values = V_total(r, q1, q2, A, b)

# Step 8: Plot the potentials
plt.figure(figsize=(10, 6))
plt.plot(r, V_coulomb_values, label='Coulomb Potential', color='blue')
plt.plot(r, V_lennard_jones_values, label='Lennard-Jones Potential', color='orange')
plt.plot(r, V_total_values, label='Total Potential Energy', color='green')
plt.title('Figure A: Potential Energy vs. Distance')
plt.xlabel('Distance (r)')
plt.ylabel('Potential Energy (V)')
plt.legend()
plt.grid()
plt.xlim(1.2, 3)
plt.ylim(-3, 1)  # Adjusted limits for better visibility
plt.show()

### Part 2B
print('\n ### Part 2B \n')
# Constants
q1 = 2        # Charge of ion 1
q2 = -1       # Charge of ion 2
A = 0.2       # Parameter for Lennard-Jones potential
b = 1.5       # Parameter for Lennard-Jones potential

# Distance range
r = np.linspace(1.2, 3, 100)

# Calculating the potentials
V_coulomb = (q1 * q2) / r  # Coulomb potential
V_lennard_jones = A * ((b / r) ** 12 - 2 * (b / r) ** 6)  # Lennard-Jones potential
V_total = V_coulomb + V_lennard_jones  # Total potential

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(r, V_coulomb, label='Coulomb Potential', color='blue')
plt.plot(r, V_lennard_jones, label='Lennard-Jones Potential', color='orange')
plt.plot(r, V_total, label='Total Potential Energy', color='green')
plt.title('Figure B: Potential Energy vs. Distance')
plt.xlabel('Distance (r)')
plt.ylabel('Potential Energy (V)')
plt.legend()
plt.grid()
plt.xlim(1.2, 3)
plt.ylim(-3, 1)  # Adjusted limits for better visibility
plt.show()

### Step 2C
print('\n ### Part 2C \n')
# Calculating forces (negative gradient of the total potential)
dV_total = np.gradient(V_total, r)
forces = -dV_total

# Plotting the forces
plt.figure(figsize=(10, 6))
plt.plot(r, np.abs(forces), label='Magnitude of Total Force', color='purple')
plt.title('Figure C: Magnitude of Force vs. Distance')
plt.xlabel('Distance (r)')
plt.ylabel('Magnitude of Force |F|')
plt.axhline(0, color='black', linewidth=0.5, ls='--')  # Add a horizontal line at zero
plt.legend()
plt.grid()
plt.xlim(1.2, 3)
plt.ylim(0, 3)  # Adjusted limits for better visibility
plt.show()

# Finding the magnitude of force at the estimated ideal distance
ideal_distance = 2.0  # Replace with your estimated distance from part C
ideal_force_magnitude = np.abs(np.interp(ideal_distance, r, forces))
print(f'Magnitude of total force at ideal distance ({ideal_distance}): {ideal_force_magnitude}')

### Part 2D
print('\n ### Part 2D \n')
# Constants
q1 = 2        # Charge of ion 1
q2 = -1       # Charge of ion 2
A = 0.2       # Parameter for Lennard-Jones potential
b = 1.5       # Parameter for Lennard-Jones potential

# Distance range
r = np.linspace(1.2, 3, 100)

# Calculate the Coulomb potential
def V_coulomb(r, q1, q2):
    return (q1 * q2) / r

# Calculate the Lennard-Jones potential
def V_lennard_jones(r, A, b):
    return A * ((b / r) ** 12 - 2 * (b / r) ** 6)

# Total potential energy
V_total = V_coulomb(r, q1, q2) + V_lennard_jones(r, A, b)

# Calculate the forces (negative gradient of the total potential)
dV_total = np.gradient(V_total, r)
forces = -dV_total

# Plotting the magnitude of forces
plt.figure(figsize=(10, 6))
plt.plot(r, np.abs(forces), label='Magnitude of Total Force', color='purple')
plt.title('Figure D: Magnitude of Force vs. Distance')
plt.xlabel('Distance (r)')
plt.ylabel('Magnitude of Force |F|')
plt.axhline(0, color='black', linewidth=0.5, ls='--')  # Add a horizontal line at zero
plt.legend()
plt.grid()
plt.xlim(1.2, 3)
plt.ylim(0, 3)  # Adjusted limits for better visibility
plt.show()

# Finding the magnitude of force at the ideal distance from part C
ideal_distance = 2.0  # Replace with your estimated ideal distance
ideal_force_magnitude = np.abs(np.interp(ideal_distance, r, forces))
print(f'Magnitude of total force at ideal distance ({ideal_distance}): {ideal_force_magnitude}')

### Part 2E
print('\n ### Part 2E \n')
# Define the charges
q1 = 2  # Charge of ion 1
q2 = -1  # Charge of ion 2
q3 = 1  # Charge of ion 3

# Define distances between charges
r = 2.0   # Distance between q1 and q2
d1 = 1.5  # Distance between q1 and q3
d2 = 2.5  # Distance between q2 and q3

# Calculate the total potential energy
V_total = (q1 * q2) / r + (q1 * q3) / d1 + (q2 * q3) / d2

# Output the Derivation of Coulomb's
print('Derivation of Coulomb Interaction (q1 * q2) / r + (q1 * q3) / d1 + (q2 * q3) / d2')

# Output the result
print('Total Potential Energy: {:.2f}'.format(V_total))
```

```python
import matplotlib.pyplot as plt

# Function to read phi-psi angles from a text file
def read_phi_psi_angles(base_path):
    angles = {}
    current_protein = None

    with open(os.path.join(base_path, "phi_psi_angles.txt"), 'r') as file:
        for line in file:
            line = line.strip()
            if line.startswith('Protein ='):
                current_protein = line.split('=')[1].strip()
                angles[current_protein] = {'phi': [], 'psi': []}
            elif current_protein and ':' in line:
                # Extract phi and psi angles from the line
                parts = line.split(':')
                angle_values = parts[1].strip().strip('()').split(',')
                phi = float(angle_values[0].strip())
                psi = float(angle_values[1].strip())
                angles[current_protein]['phi'].append(phi)
                angles[current_protein]['psi'].append(psi)

    return angles

# Function to plot phi-psi angles
def plot_phi_psi(angles):
    plt.figure(figsize=(12, 6))
    for protein, angle_data in angles.items():
        plt.scatter(angle_data['phi'], angle_data['psi'], label=protein)
    plt.title('Phi-Psi Angles')
    plt.xlabel('Phi')
    plt.ylabel('Psi')
    plt.axhline(0, color='gray', lw=0.5, ls='--')
    plt.axvline(0, color='gray', lw=0.5, ls='--')
    plt.xlim(-180, 180)
    plt.ylim(-180, 180)
    plt.grid()
    plt.legend()
    plt.show()

# Read angles and plot
phi_psi_angles = read_phi_psi_angles(base_path)
plot_phi_psi(phi_psi_angles)
```
